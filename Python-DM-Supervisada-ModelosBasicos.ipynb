# Tutorial Python: Modelos Basicos de Prediccion

Importemos los libraries necesarios para estimar todos los modelos y evaluarlos 

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import scale
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

Leer datos de una hoja de CSV y mostrar primeras observaciones

```python
datos = pd.read_csv('/Users/carloseqa/Dropbox/RStudioProjects/PredictiveAnalytics/energy.csv')
datos
```

Reportar estadisticos sobre la hoja de datos 

```python
datos.describe()
```

Definir variable dependiente 'y' como la variable HeatingLoad.
Definir variables dependientes 'X' eliminando las columnas 2, 9 y 10 (en Python 1,8 y 9; Python comienza a contar del 0)
Normalizar los datos para cuando estimemos KNN. 


```python
y=datos.HeatingLoad
X=datos.drop(datos.columns[[1,8,9]],axis=1)
X_scaled = scale(X)
```

Especificar que evaluacion se hara con 10-fold Cross-validation

```python
k_fold = KFold(n_splits=10)
```

Instanciar primer model, Regresion Lineal. Obtener MSE para X-validation. Tomar raiz cuadarada a MSE para obtener RMSE y promediar.

```python
model1 = LinearRegression()
scores1= cross_val_score(model1, X, y, cv=k_fold, n_jobs=-1,scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores1))
```

Instanciar segundo modelo, Arboles de Regresion con una profundidad de 10. Obtener MSE para X-validation. Tomar raiz cuadarada a MSE para obtener RMSE y promediar.

```python
model2 = DecisionTreeRegressor(random_state=0,max_depth=10)
scores2= cross_val_score(model2, X, y, cv=k_fold, n_jobs=-1,scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores2))
```

Instaciar tercer modelo, KNN con NumeroVecinos = 30. Obtener MSE para X-validation. Tomar raiz cuadarada a MSE para obtener RMSE y promediar.

```python
model3 = KNeighborsRegressor(n_neighbors=30)
scores3= cross_val_score(model3, X_scaled, y, cv=k_fold, n_jobs=-1,scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores3))
```

Instanciar cuarto modelo, KNN sin especificar NumeroVecinos. Hacer busqueda de 'n_neighbor' optimo. Obtener MSE para X-validation. Tomar raiz cuadarada a MSE para obtener RMSE y promediar.

```python
pre_model4 = KNeighborsRegressor()
parameters = {'n_neighbors':[1, 5,10,30,100]}
model4 = GridSearchCV(pre_model4, parameters)
scores4= cross_val_score(model4, X_scaled, y, cv=k_fold, n_jobs=-1,scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores4))
```
