# Tutorial Python: Leyendo Hojas CSV y Explorandolas.

Primero importemos la libraries necesarias para importar las hojas de CSV y para hacer los graficos.

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

Importemos la hoja HousePrices.csv y guardemosla en el objeto 'datos'. Examinemos las primeras observaciones de datos y despues calculemos estadisticos descriptivos.

```python
datos = pd.read_csv('HousePrices.csv')
```

```python
datos.head()
```

```python
datos.describe()
```

## Una Sola Variable

### Variables Continuas

Primero vamos a dibujar un histograma para las variables continuas. Primero **SqFt** y despues **Price**. Experimenten cambiando 'kde' y 'rug' a True.

```python
sns.distplot(datos.Price,kde=False,hist=True,rug=False)
```

```python
sns.distplot(datos.SqFt,kde=True,hist=True,rug=False)
```

Ahora probemos boxplots. Primero **Price** y despues **SqFt**.

```python
sns.boxplot(x="Price",data=datos,orient='v')
```

```python
sns.boxplot(x="SqFt",data=datos,orient='v')
```

### Variables Categoricas

Barras son una de las mejores maneras de representar conteos. Primero **Brick** y despues **Neighborhood**.

```python
sns.countplot(x="Brick", data=datos)
```

```python
sns.countplot(x="Neighborhood", data=datos)
```

## Dos Variables

### Continua-Continua

Scatter plots son sin duda la mejor opcion. Podemos incluir la recta de regresion.

```python
sns.regplot(x="SqFt",y="Price",data=datos,fit_reg=False)
```

```python
sns.regplot(x="SqFt",y="Price",data=datos,fit_reg=True)
```

Una variacion del grafico incluye histogramas de las variables en el scatter plot.

```python
sns.jointplot(x="SqFt",y="Price",data=datos)
```

Podemos dibujar boxplots por categoria. Primero para **Neighborhood** y despues para **Brick**.

```python
sns.boxplot(x="Neighborhood", y="Price", data=datos)
```

```python
sns.boxplot(x="Brick", y="Price", data=datos)
```

Una variacion es el diagrama de violin. 

```python
sns.violinplot(x="Neighborhood", y="Price", data=datos)
```

```python
sns.violinplot(x="Brick", y="Price", data=datos)
```

Otra variacion son barras en las que la altura de la barra refleja la media del precio.

```python
sns.barplot(x="Neighborhood", y="Price", data=datos,ci=None)
```

```python
sns.barplot(x="Brick", y="Price", data=datos,ci=None)
```
