# Tutorial Python: Modelos Basicos de Clasificacion

Cargar las libraries utility

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import scale
from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV
```

Cargar las libraries para estimar los distintos algoritmos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

Set up X-validation (version estratificada)

```python
k_fold = StratifiedKFold(n_splits=10)
```

Leer datos y echarles un vistazo a primeras observaciones

```python
datos = pd.read_csv('/Users/carloseqa/Dropbox/RStudioProjects/PredictiveAnalytics/bwt.csv')
datos.head()
```

Sklearn solo acepta variables numericas (aun para clasificacion). Es necesario definir dummies para todas las variables categoricas, incluyendo la variable dependiente. Python creara dummies para todas las categorias; una de ellas siempre es redundante. El setting 'drop_first=True' elimina la primera categoria. 

```python
datos=pd.get_dummies(datos,drop_first=True)
datos.head()
```

Sklearn pide variable dependiente e independientes por separado. Normalicemos desde ahora los datos para uso con KNN. 

```python
y = datos.lowweight_yes
X = datos.drop(datos.columns[[0,10]],axis=1)
X_scaled = scale(X)
X.head()
```

### Regresion Logistica

Ahora instanciemos y estimemos el modelo de Regresion Logistica. Scikit-learn, por default, trata de hacer regularizacion de los coeficientes. Asignando a C un valor altisimo se evita la regularizacion. 

```python
modelo1 = LogisticRegression(C=1e25)
modelo1.fit(X,y)
```

```python
X.columns
```

Echemosle un vistazo a los coeficientes y al intercepto. 

```python
print(modelo1.coef_) 
```

```python
print(modelo1.intercept_)
```

Ahora evaluemos performance (accuracy)

```python
scores1= cross_val_score(modelo1, X, y, cv=k_fold, n_jobs=-1)
np.mean(scores1)
```

Clasifiquemos las observaciones y despues obtengamos probabilidades de exito y fracaso 

```python
pred1 = modelo1.predict(X)
pred1
```

```python
pred1 = modelo1.predict_proba(X)
pred1
```

### Arboles de Clasificacion 

```python
model2 = DecisionTreeClassifier(random_state=0,max_depth=10)
model2.fit(X,y)
```

```python
scores2= cross_val_score(model2, X, y, cv=k_fold, n_jobs=-1)
np.mean(scores2)
```

Los arboles de Scikit-learn son de alta calidad pero requieren importar algunas libraries adicionales.

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(model2, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
